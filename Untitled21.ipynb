```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
import string
import re
import tensorflow as tf
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
data =pd.read_csv(r'arts.csv')
```

```python
data['headline'].values
```

```python
def clean_text(line):
    newline = "".join([w for w in line if w not in string.punctuation]).lower()
    doc = nlp(newline)
    
    word=[]
    for token in list(doc):
        if token.text==' ':
            continue
        else:
            word.append(token.text)
    return " ".join(word)
```

```python
headlines=[]
for line in data['headline'].values:
    cleaned_text=clean_text(line)+' eos'
    headlines.append(cleaned_text)
```

```python
headlines
```

```python
t  = tf.keras.preprocessing.text.Tokenizer()
t.fit_on_texts(headlines)
sequences = t.texts_to_sequences(headlines)
```

```python
xdata=[]
ydata=[]
for seq in sequences:
    seqx=seq[:-1]
    seqy=seq[1:]
    xdata.append(seqx)
    ydata.append(seqy)
```

```python
pad_seqx=tf.keras.preprocessing.sequence.pad_sequences(xdata, padding='post',maxlen=25)
pad_seqy=tf.keras.preprocessing.sequence.pad_sequences(ydata, padding='post',maxlen=25)
```

```python
hot_pad_y=tf.keras.utils.to_categorical(pad_seqy)
```

```python
input= tf.keras.Input((25))
embd=tf.keras.layers.Embedding(len(t.word_index)+1,3,input_length=25)(input)
l=tf.keras.layers.LSTM(100,return_sequences=True)(embd)
d=tf.keras.layers.TimeDistributed(tf.keras.layers.Dense(len(t.word_index)+1,activation='softmax'))(l)
model=tf.keras.Model(input,d)
model.summary()
```

```python
model.compile(loss=tf.keras.losses.categorical_crossentropy,optimizer=
              tf.keras.optimizers.Adam(0.01))
```

```python
model.fit(pad_seqx,hot_pad_y,epochs=1000,batch_size=32)
```

```python
text="play the americans"
ctext=t.texts_to_sequences([text])
padt=tf.keras.preprocessing.sequence.pad_sequences(ctext, padding='post',maxlen=25).astype(np.float32)
pred=model.predict(padt)
print(np.argmax(pred[0,2]))
```

```python
text="will"
i=(len(text.split())-1)
while True:
  ctext=t.texts_to_sequences([text])
  padt=tf.keras.preprocessing.sequence.pad_sequences(ctext, padding='post',maxlen=25).astype(np.float32)
  pred=model.predict(padt)
  n=np.argmax(pred[0,i])
  if n == 1 or n == 0:  #==eos
    break
  else:
    word=list(t.word_index.keys())[list(t.word_index.values()).index(n)]
    text+=" "+word
    i+=1

print(text)
```
